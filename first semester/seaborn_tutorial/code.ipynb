```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
#distribution plot
sns.displot(crash_df['not_distracted'], kde = True, bins = 10)
```

```python
#joint plot
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
#kde plot
sns.kdeplot(crash_df['alcohol'])
```

```python
#pair plot
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
#rug plot
sns.rugplot(tips_df['tip'])
```

Styling

```python
sns.set_style('ticks')

plt.figure(figsize = (10,10))

#size of display
sns.set_context('paper', font_scale = 1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

#deleting the axis
sns.despine(left=True)
```

Categorical data

```python
#barplot
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
```

```python
#countplot
sns.countplot(x='sex', data=tips_df)
```

```python
#box plots
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
#violin plot
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
#dodge - seperating data filtered by hue
#jitter - separating the points to not let them be so stacked on themselves
```

```python
#swarm plot is just like stripplot, but the points are adjusted to not overlap, it is some kind of a combination between stripplot and boxplot
sns.violinplot(x='day', y='total_bill', data=tips_df)
#stacking swarmplot on violinplot
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
crash_mx = crash_df.corr()
sns.heatmap(crash_mx, annot=True)
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights, cmap='Blues')
```

```python
iris = sns.load_dataset('iris')
```

```python
#cluster maps
species = iris.pop('species')

sns.clustermap(iris)
```

```python
#pair grid
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
iris_g = sns.PairGrid(iris, hue='species',
                    x_vars=['sepal_length', 'sepal_width'],
                    y_vars=['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
#facet grid
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker',
                        col_order=['Dinner', 'Lunch'], 
                        aspect=1.3, 
                        height=4)
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
#regression plots
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips_df)
```
